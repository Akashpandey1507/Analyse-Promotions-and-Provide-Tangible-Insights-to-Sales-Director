```python
import pandas as pd
import requests
from io import BytesIO
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
url = "https://github.com/Akashpandey1507/Analyse-Promotions-and-Provide-Tangible-Insights-to-Sales-Director/raw/7b239407b70dd844cde733efb159fd04d60a5776/cleaned-data/cleaned-data"
```

```python
# Download the file content
response = requests.get(url)
parquet_file = BytesIO(response.content)
parquet_file
```

```python
# Read the Parquet file into a DataFrame
df = pd.read_parquet(parquet_file)
```
